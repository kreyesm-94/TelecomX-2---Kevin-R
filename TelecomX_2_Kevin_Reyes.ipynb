# RealizaciÃ³n de Encoding #

Transforma las variables categÃ³ricas a formato numÃ©rico para hacerlas compatibles con los algoritmos de machine learning. Utiliza un mÃ©todo de codificaciÃ³n adecuado, como one-hot encoding.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


cat_cols = df_clean.select_dtypes(include=['object']).columns
num_cols = df_clean.select_dtypes(exclude=['object']).drop(columns=['Churn']).columns


preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', sparse_output=False), cat_cols),
        ('num', 'passthrough', num_cols)
    ]
)
df_encoded = preprocessor.fit_transform(df_clean.drop(columns=['Churn']))

print("Shape despuÃ©s de codificaciÃ³n:", df_encoded.shape)
```

# SeparaciÃ³n de Datos #

Se devide el conjunto de datos en entrenamiento y prueba para evaluar el rendimiento del modelo. Una divisiÃ³n comÃºn es 70% para entrenamiento y 30% para prueba, o 80/20, dependiendo del tamaÃ±o de la base de datos.



```python
from sklearn.model_selection import train_test_split

# DivisiÃ³n en train/test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y_smote,
    test_size=0.3,
    random_state=42,
    stratify=y_smote
)

print("TamaÃ±o entrenamiento:", X_train.shape, y_train.shape)
print("TamaÃ±o prueba:", X_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Escalador
scaler = StandardScaler()

# Ajusta y transforma solo las variables de entrenamiento
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train_scaled, y_train)

y_pred_log = log_model.predict(X_test_scaled)
```

# CreaciÃ³n de Modelos #

 La decisiÃ³n de aplicar o no la normalizaciÃ³n depende de los modelos seleccionados. Ambos modelos pueden ser creados sin normalizaciÃ³n, pero tambiÃ©n es una opciÃ³n combinar modelos con y sin normalizaciÃ³n.

* Creamos un modelo basado en distancia (la RegresiÃ³n LogÃ­stica, que necesita normalizaciÃ³n â†’ usamos StandardScaler).

* Creamos un modelo basado en Ã¡rboles (el Random Forest, que no requiere normalizaciÃ³n).


```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# ========================
# 1. MODELO LOGÃSTICO
# ========================
log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train_scaled, y_train)

y_pred_log = log_model.predict(X_test_scaled)

print("Rendimiento - RegresiÃ³n LogÃ­stica")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))

# ========================
# 2. RANDOM FOREST
# ========================
rf_model = RandomForestClassifier(n_estimators=200, random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("Rendimiento - Random Forest")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))

# ========================
# 3. MATRICES DE CONFUSIÃ“N
# ========================
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.heatmap(confusion_matrix(y_test, y_pred_log), annot=True, fmt='d', cmap="Blues", ax=axes[0])
axes[0].set_title("Matriz de confusiÃ³n - LogÃ­stica")
axes[0].set_xlabel("PredicciÃ³n")
axes[0].set_ylabel("Real")

sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt='d', cmap="Greens", ax=axes[1])
axes[1].set_title("Matriz de confusiÃ³n - Random Forest")
axes[1].set_xlabel("PredicciÃ³n")
axes[1].set_ylabel("Real")

plt.tight_layout()
plt.show()
```

# **JustificaciÃ³n del preprocesamiento**

Para la RegresiÃ³n LogÃ­stica, aplicamos normalizaciÃ³n porque es un modelo basado en optimizaciÃ³n matemÃ¡tica (gradiente descendente). Si no escalamos, variables con rangos grandes podrÃ­an dominar el entrenamiento.

Para Random Forest, no aplicamos normalizaciÃ³n porque los Ã¡rboles se basan en umbrales de particiÃ³n (ej. â€œedad < 30â€), lo que hace irrelevante la escala de la variable.

# **ComparaciÃ³n de resultados**

La RegresiÃ³n LogÃ­stica obtuvo una accuracy de ~0.85 y un buen recall en la clase positiva (clientes que cancelan).

El Random Forest superÃ³ ligeramente con accuracy de ~0.87, mostrando mayor balance entre precisiÃ³n y recall.

# **ConclusiÃ³n del paso**

Ambos modelos son vÃ¡lidos:

La RegresiÃ³n LogÃ­stica es Ãºtil si queremos interpretabilidad (coeficientes que muestran impacto de cada variable).
