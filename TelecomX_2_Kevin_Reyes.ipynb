# Realización de Encoding #

Transforma las variables categóricas a formato numérico para hacerlas compatibles con los algoritmos de machine learning. Utiliza un método de codificación adecuado, como one-hot encoding.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


cat_cols = df_clean.select_dtypes(include=['object']).columns
num_cols = df_clean.select_dtypes(exclude=['object']).drop(columns=['Churn']).columns


preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', sparse_output=False), cat_cols),
        ('num', 'passthrough', num_cols)
    ]
)
df_encoded = preprocessor.fit_transform(df_clean.drop(columns=['Churn']))

print("Shape después de codificación:", df_encoded.shape)
```

# Separación de Datos #

Se devide el conjunto de datos en entrenamiento y prueba para evaluar el rendimiento del modelo. Una división común es 70% para entrenamiento y 30% para prueba, o 80/20, dependiendo del tamaño de la base de datos.



```python
from sklearn.model_selection import train_test_split

# División en train/test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y_smote,
    test_size=0.3,
    random_state=42,
    stratify=y_smote
)

print("Tamaño entrenamiento:", X_train.shape, y_train.shape)
print("Tamaño prueba:", X_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Escalador
scaler = StandardScaler()

# Ajusta y transforma solo las variables de entrenamiento
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train_scaled, y_train)

y_pred_log = log_model.predict(X_test_scaled)
```

# Creación de Modelos #

 La decisión de aplicar o no la normalización depende de los modelos seleccionados. Ambos modelos pueden ser creados sin normalización, pero también es una opción combinar modelos con y sin normalización.

* Creamos un modelo basado en distancia (la Regresión Logística, que necesita normalización → usamos StandardScaler).

* Creamos un modelo basado en árboles (el Random Forest, que no requiere normalización).


```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# ========================
# 1. MODELO LOGÍSTICO
# ========================
log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train_scaled, y_train)

y_pred_log = log_model.predict(X_test_scaled)

print("Rendimiento - Regresión Logística")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))

# ========================
# 2. RANDOM FOREST
# ========================
rf_model = RandomForestClassifier(n_estimators=200, random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("Rendimiento - Random Forest")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))

# ========================
# 3. MATRICES DE CONFUSIÓN
# ========================
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.heatmap(confusion_matrix(y_test, y_pred_log), annot=True, fmt='d', cmap="Blues", ax=axes[0])
axes[0].set_title("Matriz de confusión - Logística")
axes[0].set_xlabel("Predicción")
axes[0].set_ylabel("Real")

sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt='d', cmap="Greens", ax=axes[1])
axes[1].set_title("Matriz de confusión - Random Forest")
axes[1].set_xlabel("Predicción")
axes[1].set_ylabel("Real")

plt.tight_layout()
plt.show()
```

# **Justificación del preprocesamiento**

Para la Regresión Logística, aplicamos normalización porque es un modelo basado en optimización matemática (gradiente descendente). Si no escalamos, variables con rangos grandes podrían dominar el entrenamiento.

Para Random Forest, no aplicamos normalización porque los árboles se basan en umbrales de partición (ej. “edad < 30”), lo que hace irrelevante la escala de la variable.

# **Comparación de resultados**

La Regresión Logística obtuvo una accuracy de ~0.85 y un buen recall en la clase positiva (clientes que cancelan).

El Random Forest superó ligeramente con accuracy de ~0.87, mostrando mayor balance entre precisión y recall.

# **Conclusión del paso**

Ambos modelos son válidos:

La Regresión Logística es útil si queremos interpretabilidad (coeficientes que muestran impacto de cada variable).

El Random Forest es mejor si buscamos mayor rendimiento predictivo y podemos sacrificar un poco de interpretabilidad.

**Regresión Logística:** No parece overfit (su desempeño es consistente), pero al ser lineal puede estar subajustando ciertas relaciones complejas → ligero underfitting.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc
```

```python
# CURVAS ROC

y_prob_log = log_model.predict_proba(X_test_scaled)[:, 1]
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]

fpr_log, tpr_log, _ = roc_curve(y_test, y_prob_log)
roc_auc_log = auc(fpr_log, tpr_log)

fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)

plt.figure(figsize=(8,6))
plt.plot(fpr_log, tpr_log, color="blue", label=f"Logística (AUC = {roc_auc_log:.2f})")
plt.plot(fpr_rf, tpr_rf, color="green", label=f"Random Forest (AUC = {roc_auc_rf:.2f})")
plt.plot([0,1], [0,1], color="gray", linestyle="--")

plt.title("Curvas ROC")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

Se realiza un gráfico con la curva ROC tomando en consideración cambos modelos y resultados optenidos

## Estrategias de retención recomendadas

1. **Sensibilidad a precios (Charges.Total / Charges.Monthly):**
   - Ofrecer **planes escalonados** y **descuentos por permanencia** a clientes con cargos elevados.
   - Implementar **alertas proactivas** cuando la factura supere umbrales históricos del cliente.

2. **Tipo de servicio (InternetService_Fiber optic):**
   - Monitorear **incidencias y velocidad percibida** en fibra; activar **task force de calidad** en zonas con mayor churn.
   - Paquetes de **valor agregado** (mejor router, instalación premium) con compromiso mínimo flexible.

3. **Servicios de entretenimiento (StreamingTV / StreamingMovies):**
   - Bundles con **precio preferencial** o **meses gratis** condicionados a permanencia.
   - **Recomendaciones personalizadas**: detectar poco uso y proponer **downgrade asistido** antes de que el cliente cancele todo.

4. **PaperlessBilling y método de pago (Electronic check):**
   - Incentivar **autopago** con tarjeta/cuenta y **factura electrónica con beneficios** (cashback, puntos, comisiones $0).
   - Campañas de **educación financiera** y recordatorios para reducir fricción en pago.

5. **Antigüedad (tenure):**
   - **Programa “primeros 90 días”**: check-ins proactivos, encuesta NPS temprana y resolución rápida de tickets.
   - **Bonos de lealtad** tierizados por hitos de permanencia (6, 12, 24 meses).
